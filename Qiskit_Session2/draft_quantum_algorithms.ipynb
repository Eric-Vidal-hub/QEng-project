### Constant oracle $U_{f_1}$

Which is not operating on bits.

````python
# Deustch-Josza's Algorithm
n=4
qc=QuantumCircuit(n+1,n)
for i in range(n):
    qc.h(i)

# Start the |-> state
qc.x(n)
qc.h(n)
qc.barrier()

# Oracle
# Nothing to apply
qc.barrier()

# Inversion
for i in range(n):
    qc.h(i)
qc.barrier()

qc.measure(range(n),range(n))
qc.draw(output='mpl')
````

````python
# Results
job=execute(qc,backend=simulator,shots=1024)
result=job.result()
counts=result.get_counts(qc)
print(counts)
````

### Balanced oracle $U_{f_2}$

The one done in the past section to get ready.

````python
# Deustch-Josza's Algorithm
n=4
qc=QuantumCircuit(n+1,n)
for i in range(n):
    qc.h(i)

# Start the |-> state
qc.x(n)
qc.h(n)
qc.barrier()

# Oracle
for i in range(n):
    qc.cx(i,n)
qc.barrier()

# Inversion
for i in range(n):
    qc.h(i)
qc.barrier()

qc.measure(range(n),range(n))
qc.draw(output='mpl')
````

````python
# Results
job=execute(qc,backend=simulator,shots=1024)
result=job.result()
counts=result.get_counts(qc)
print(counts)
````

## Bernstein-Vazirani

Bernstein-Vazirani algorithm is an algorithm that solve the following problem:
Let $u=u_{n-1}\dots u_0\in \{0,1\}^n$ an unknown $n$-bits string. Suppose you have a classical function $f:\{0,1\}^n\to \{0,1\}$ which calculates $f_u(x)=x.u [2]$ where $x.u$ is the usual inner product between $n$-bits strings, i.e. $x=x_{n-1}\dots x_0$ and $$x.u=x_0u_0+x_1u_1+\dots+x_{n-1}u_{n-1}$$

The goal of Bernstein-Vazirani's algorithm is to find $u$.

Recall that classically you can find $u$ by testing $n$ values $x\in \{0,1\}^n$. Bernstein-Vazirani's algorithm solves the problem by using only once the function $f$.
